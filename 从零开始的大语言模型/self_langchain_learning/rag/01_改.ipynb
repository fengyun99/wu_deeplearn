```python
# 加载环境
from dotenv import load_dotenv,find_dotenv
_ = load_dotenv(find_dotenv())
```

```python
# 加载文件
from langchain_community.document_loaders import DirectoryLoader,UnstructuredFileLoader
loader = DirectoryLoader('../data',
                        glob="**/*.docx",
                        use_multithreading=True,
                        loader_cls=UnstructuredFileLoader).load()
```

```python
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.vectorstores import ElasticVectorSearch
from langchain_openai import OpenAIEmbeddings

text_splitter = CharacterTextSplitter(chunk_size=400, chunk_overlap=100)
texts = text_splitter.split_documents(loader)
embeddings = OpenAIEmbeddings()

db = ElasticVectorSearch.from_documents(texts, embeddings)
```
